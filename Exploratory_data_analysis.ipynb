```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
titanic_data = pd.read_csv('test.csv')

# Display the first few rows of the dataset
titanic_data.head()
```

```python
# Check for missing values
titanic_data.isnull().sum()
```

```python
# Fill missing 'Age' with median
titanic_data['Age'] = titanic_data['Age'].fillna(titanic_data['Age'].median())

# Drop 'Cabin' column (too many missing values)
titanic_data.drop(columns=['Cabin'], inplace=True)
```

```python
# Check for missing values
titanic_data.isnull().sum()
```

```python
# Display basic statistics
titanic_data.describe()
```

```python
# Histogram of Age
sns.histplot(titanic_data['Age'], bins=20, kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()          # Age distribution suggests most passengers were between 20–40 years old.
```

```python
# Heatmap of correlations
plt.figure(figsize=(10, 6))
sns.heatmap(titanic_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
